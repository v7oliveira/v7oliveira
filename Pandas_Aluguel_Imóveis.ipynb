## Imóveis Residenciais

```python
dados.head(10)
```

```python
## filtrar os dados para que apresente Imóveis com a categoria "Residenciais"

list(dados['Tipo'].drop_duplicates())
```

```python
## utilizar os dados extraidos e filtrar por tipos que se encaixam em "residencial"

residencial = ['Quitinete',
 'Casa',
 'Apartamento',
 'Casa de Condomínio',
 'Casa Comercial',
 'Casa de Vila']

residencial
```

```python
selecao = dados['Tipo'].isin(residencial)

selecao
```

```python
## utilizando o método de seleção de DataFrame

dados_residencial = dados[selecao]
dados_residencial
```

```python
list(dados_residencial['Tipo'].drop_duplicates())
```

```python
## fazer Upload do arquivo filtrado apenas com dados residenciais

from google.colab import files
uploaded = files.upload()
```

```python
dados_2 = pd.read_csv("dados_residenciais.csv", sep = ",")

dados_2
```

```python
# Selecionar somente os imóveis classificados como tipo 'Apartamento'

selecao = dados_2['Tipo'] == 'Apartamento'
n1 = dados_2[selecao].shape[0]
n1
```

```python
# Selecione os imóveis classificados com tipos 'Casa', 'Casa de condominio' e 'Casa de vila'

selecao = (dados_2['Tipo'] == 'Casa') | (dados_2['Tipo'] == 'Casa de Condomínio') | (dados_2['Tipo'] == 'Casa de Vila')

n2 = dados_2[selecao].shape[0]

n2
```

```python
# Selecionar os imóveis com área entre 60 e 100 metros quadrados, incluindo limites

selecao = (dados_2['Area'] >= 60) & (dados_2['Area'] <=100)
n3 = dados_2[selecao].shape[0]
n3
```

```python
# Selecionar os imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00

selecao = (dados_2['Quartos'] >= 4) & (dados_2['Valor'] < 2000)
n4 = dados_2[selecao].shape[0]
n4
```

```python
print(" Nº de imóveis classificados como tipo 'Apartamento' -> {}".format(n1))
print(" Nº de imóveis classificados com tipos 'Casa', 'Casa de condominio' e 'Casa de vila' -> {}".format(n2))
print(" Nº de imóveis com área entre 60 e 100 metros quadrados, incluindo limites -> {}".format(n3))
print(" Nº de imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00 -> {}".format(n4))
```
